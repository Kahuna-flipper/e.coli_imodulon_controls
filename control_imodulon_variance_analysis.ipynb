```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os
path = os.getcwd()
```

```python
metadata = pd.read_excel(path+'/Data/'+'metadata.xlsx',sheet_name = 1)
imodulon_annotations = pd.read_excel(path+'/Data/'+'imodulon_annotations.xlsx',sheet_name=0)
```

```python
microarray_metadata = metadata[(metadata['Dataset']=='MA-1') | (metadata['Dataset']=='MA-2')  | (metadata['Dataset']=='MA-3')]
microarray_metadata
```

```python
microarray_activity_1 = pd.read_excel(path+'/Data/'+'activities.xlsx',sheet_name = 3,index_col='Unnamed: 0')
microarray_activity_2 = pd.read_excel(path+'/Data/'+'activities.xlsx',sheet_name = 4,index_col='Unnamed: 0')
microarray_activity_3 = pd.read_excel(path+'/Data/'+'activities.xlsx',sheet_name = 5,index_col='Unnamed: 0')


controls_1 = []
controls_2 = []
controls_3 = []

for column in microarray_activity_1.columns:
    if('N0000' in column):
        controls_1.append(column)
        
for column in microarray_activity_2.columns:
    if('aero' in column and 'anaero' not in column):
        controls_2.append(column)
        
for column in microarray_activity_3.columns:
    if('glucose' in column):
        controls_3.append(column)
```

```python
microarray_activity_1[controls_1]
```

```python
variances_MA1 = microarray_activity_1[controls_1].var(axis=1).sort_values(ascending = False)
variances_MA2 = microarray_activity_2[controls_2].var(axis=1).sort_values(ascending = False)
variances_MA3 = microarray_activity_3[controls_3].var(axis=1).sort_values(ascending = False)
```
